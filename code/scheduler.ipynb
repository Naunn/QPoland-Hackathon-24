```python
import pandas as pd
import numpy as np

import os

PATH = os.getcwd().replace("code", "")
Simpler_instance = PATH + "data\\Simpler_instance\\"
```

```python
# Prosty harmonogram lotów bez użycia obiektów

# Lista przechowująca loty
flights = []
aircrafts = []


# Funkcja do dodawania lotu
def add_flight(
    leg_id: int,
    flight_id: int,
    aircraft: str,
    origin: str,
    destination: str,
    departure_time,
    arrival_time,
) -> None:
    flight = {
        "leg_id": leg_id,
        "flight_id": flight_id,
        "aircraft": aircraft,
        "origin": origin,
        "destination": destination,
        "departure_time": departure_time,
        "arrival_time": arrival_time,
    }
    flights.append(flight)


def add_aircraft(
    registration: str, model: str, family: str, position: str, owned: int
) -> None:
    aircraft = {
        "registration": registration,
        "model": model,
        "family": family,
        "position": position,
        "owned": owned,
    }
    aircrafts.append(aircraft)


# Funkcja do wyświetlania wszystkich lotów
def display_flights() -> None:
    if not flights:
        print("No flights available.")
    else:
        for flight in flights:
            print(
                f"Leg identifier {flight['leg_id']} Flight {flight['flight_id']} from {flight['origin']} to {flight['destination']}, "
                f"departing at {flight['departure_time']}, arriving at {flight['arrival_time']}"
            )


# Funkcja do wyświetlania dostępnej floty
def display_aircrafts() -> None:
    if not aircrafts:
        print("No aircrafts available.")
    else:
        for aircraft in aircrafts:
            print(aircraft)


# Funkcja do znajdowania dostępnych samolotów na podstawie lokalizacji
def find_aircraft_by_location(position: str) -> list:
    found_aircrafts = [
        aircraft for aircraft in aircrafts if aircraft["position"] == position
    ]
    if not found_aircrafts:
        print(f"No fleet in {position} available.")
    else:
        for aircraft in found_aircrafts:
            print(aircraft)

    return found_aircrafts


# Funkcja do usuwania lotu
def remove_flight(flight_id: int) -> None:
    global flights
    flights = [flight for flight in flights if flight["flight_id"] != flight_id]
    print(f"Flight {flight_id} has been removed.")


# Przykładowe użycie

# Dodawanie floty
add_aircraft("A321#4", "A321", "Airbus", "CDG", 1)
add_aircraft("A320#10", "A320", "Airbus", "TLS", 1)

# Wyświetlanie floty
print("All available aircrafts:")
display_aircrafts()

# Dodawanie lotów
add_flight(
    "70002970", "2970", "A321#4", "TLS", "ORY", "2006-01-07T06:40", "2006-01-07T08:00"
)

# Wyświetlanie wszystkich lotów
print("All available flights:")
display_flights()

# Znajdowanie dostępnej floty dla CDG
print("\nAircrafts in CDG:")
find_aircraft_by_location("CDG")

# Usuwanie lotu
remove_flight("AA123")

# Wyświetlanie lotów po usunięciu
print("\nFlights after removal:")
display_flights()
```
